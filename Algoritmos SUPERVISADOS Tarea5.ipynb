```python
df=loans_dummies.copy()
models_ed = ['AdaBoost', 'BayesRidge', 'Boosting', 'DecisionTree', 'LassoReg', 'LogisticRegression','svm','RidgeClass','RF']
n_jobs = multiprocessing.cpu_count() // 2
results_ed = {
    'model' : [],
    'balanced_accuracy' : [],
    'time' : []
}
for i in tqdm(range(30)):
    df_train = df.sample(frac = 0.7)
    x_train = df.drop("paid",axis=1)
    y_train = df[['paid']]

    df_test = df.drop(df_train.index)
    x_test = df_test.drop("paid",axis=1)
    y_test = df_test[['paid']]
    for m in models_ed:
        start = timer()
        model = fit(x_train, y_train.values.ravel(), m, n_jobs = n_jobs)
        end = timer()
        y_pred = model.predict(x_test)
        results_ed['model'].append(m)
        results_ed['balanced_accuracy'].append(balanced_accuracy_score(y_test, y_pred))
        results_ed['time'].append(end - start)

    ev = pd.DataFrame(results_ed)
    ev = ev.sort_values('balanced_accuracy')
    ev.to_csv(path+'results_comparison_supervisedalgorithms.csv',index=False)
```
