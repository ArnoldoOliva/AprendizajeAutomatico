```python
import pandas as pd
import numpy as np
```

```python
path="C:/Users/HUGO/Desktop/Maestria en CIENCIA DE DATOS UANL/Aprendizaje automatico/"
pd.set_option("display.max_columns",None)
loans=pd.read_csv(path+"SBAnational.csv")
loans
```

```python
from datetime import datetime
datescolumns=[e for e in loans.columns if "Date" in e]
for e in datescolumns:
    loans[e]=pd.to_datetime(loans[e],errors="coerce")
#loans.dtypes
```

```python
loans=loans[loans['ApprovalDate']<"2023"] #parse errors
loans.sort_values("ApprovalDate")
```

```python
loans=loans[(loans["ApprovalDate"]>"2000")&(loans["ApprovalDate"]<"2006")&
             (loans["DisbursementDate"]>"2000")&(loans["DisbursementDate"]<"2006")] #WORKING WITH LOANS WITHOUT ECONOMIC CRISIS:
loans
```

```python
#print(loans[loans["ApprovalFY"]==2006]["MIS_Status"].value_counts(),2006)
print(loans[loans["ApprovalFY"]==2005]["MIS_Status"].value_counts(),2005)
print(loans[loans["ApprovalFY"]==2004]["MIS_Status"].value_counts(),2004)
print(loans[loans["ApprovalFY"]==2003]["MIS_Status"].value_counts(),2003)
print(loans[loans["ApprovalFY"]==2002]["MIS_Status"].value_counts(),2002)
print(loans[loans["ApprovalFY"]==2001]["MIS_Status"].value_counts(),2001)
```

```python
loans["business"] = pd.factorize(loans.Name)[0] + 1
#We don´t need some specific columns (like names, geographical specific locations, etc) so we will drop them :
dropcols=["Name","City","Zip","Bank","LoanNr_ChkDgt"]
cols=[e for e in loans.columns if e not in dropcols]
loans=loans[cols]
#loans.reset_index(inplace=True)
loans["index"]=loans.index
loans.rename(columns={"index":"ID"},inplace=True)
loans
```

```python
loans["same_state"]=np.nan
loans.loc[loans["State"]==loans['BankState'],"same_state"]=1
loans.loc[loans["State"]!=loans['BankState'],"same_state"]=0

loans.drop(["State","BankState"],axis=1,inplace=True)
loans["same_state"].value_counts()
```

```python
loans.drop("ApprovalFY",axis=1,inplace=True)
loans.drop("ChgOffDate",axis=1,inplace=True)
loans["APPROVALtimestamp"] = loans.ApprovalDate.values.astype(np.int64) / 10 ** 9
loans["DISBURSEMENTtimestamp"] = loans.DisbursementDate.values.astype(np.int64) / 10 ** 9
loans["timediff"]=loans.DISBURSEMENTtimestamp-loans.APPROVALtimestamp
```

```python
print(loans["MIS_Status"].value_counts())
loans.rename(columns={"MIS_Status":"paid"},inplace=True)
loans.loc[loans["paid"]=="P I F","paid"]=1
loans.loc[loans["paid"]=="CHGOFF","paid"]=0
print(loans["paid"].value_counts())
```

```python
cols=["DisbursementGross","BalanceGross","ChgOffPrinGr","GrAppv","SBA_Appv"]
for e in cols:
    loans[e]=loans[e].str.replace("$","")
    loans[e]=loans[e].str.replace(",","")
    loans[e]=loans[e].astype(float)
    print(loans[e].dtype)
```

```python
loans.loc[loans.Term<=92.45,"Terms_comp"]=0 #bajo
loans.loc[loans.Term<=92.45,"Terms_comp"]=1 #alto
```

```python
print(loans["RevLineCr"].value_counts())
loans.loc[loans["RevLineCr"]=="0","RevLineCr"]=0
loans.loc[loans["RevLineCr"]=="Y","RevLineCr"]=1
loans.loc[(loans["RevLineCr"]=="N"),"RevLineCr"]=0
loans.loc[(~loans["RevLineCr"].isin([0,1])),"RevLineCr"]=np.nan
loans["RevLineCr"].value_counts()
```

```python
loans["LowDoc"]=loans["LowDoc"].map(lambda x: 1 if x == "Y" else 0)
loans["LowDoc"].value_counts()
```

```python
numericalcolumns=["Term","NoEmp","NewExist","CreateJob","RetainedJob","UrbanRural","RevLineCr","DisbursementGross",	"BalanceGross",	"ChgOffPrinGr",
                  "GrAppv",	"SBA_Appv","same_state","LowDoc",	"APPROVALtimestamp",	"DISBURSEMENTtimestamp",	"timediff"]
```

```python
loans["SCIAN category"] = pd.factorize(loans.NAICS)[0] + 1
loans.loc[loans['NewExist']==0,"NewExist"]=np.nan
loans.loc[loans['UrbanRural']==0,"UrbanRural"]=np.nan
```

```python
loans.to_csv(path+"loansmodified1.csv",index=False)
```
